```python
from __future__ import division

import random
import pyprind 

import egt
import species

%matplotlib inline
```

# Hawk Dove simulation with limited resources

[Example](https://www.stat.berkeley.edu/~aldous/157/Papers/haigh2.pdf)

|   |Hawk|Dove|
|---|---|---|
|**Hawk**|-2|10|
|**Dove**|0|3|

```python
print("Expected ESS at Hawks p = {}%".format(7/9*100))
```

### Run the simulation

```python
NUMBER_OF_ITERATIONS = 1000
```

##### case 1

```python
random.seed()
p = species.Population(number_of_hawks=5, number_of_doves=100)
g = egt.Game(population=p, exhaustion=10, limited_resources=True)
for _ in pyprind.prog_bar(range(0, NUMBER_OF_ITERATIONS)):
    g.advance()
    
print("\nRun complete \n")

g.population.describe()
```

```python
g.visualize('Case 1');
```
